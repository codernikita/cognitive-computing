```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
m=int(input("enter a value:"))
x=np.linspace(-10,10,20)
y=[]
z=[]
for i in x:
  y.append(m*i*i)
  z.append(m*np.sin(i))
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.grid(True)
plt.title("graph of y=mx^2 and y=msin(x)")


plt.plot(x,y,color="green",linestyle="dotted")
plt.plot(x,z,color="red",linestyle="solid")

plt.legend(["y=mx^2","y=msin(x)"])
plt.show()


```

```python
data={"c++":70,"python":90,"java":85,"html":50,"css":67}
data2=pd.DataFrame([data])
print(data2)
plt.figure(figsize=(5,5))
ax=sns.barplot(data2,palette='cividis')
plt.title("Marks in various subjects")
ax.set_xlabel("Subjects")
ax.set_ylabel("Scores")
plt.grid(True)
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()



```

```python
np.random.seed(102317245)
data=np.random.randn(50)
cum_sum=np.cumsum(data)
plt.subplot(2,2,1)
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.plot(cum_sum)
plt.title("cummulative sum")
plt.grid(True)

x=np.arange(50)
plt.subplot(2,2,2)
plt.xlabel("xaxis")
plt.ylabel("yaxis")
data2=data*0.4+0.5
plt.scatter(x,data2)
plt.title("noised data")
plt.grid(True)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
plt.subplots_adjust(wspace=0.5, hspace=0.5)



```
