```python
import numpy as np
gfg=np.matrix('4, 1, 9; 12, 3, 1; 4, 5, 6')
print(gfg)
print(f"the sum of all elements is:{np.sum(gfg)}")
print(f"the sum of elements of row is:{np.sum(gfg,axis=1)}")
print(f"the sum of elements of column is:{np.sum(gfg,axis=0)}")
```

```python
array = np.array([10, 52, 62, 16, 16, 54, 453])
print(array)
print(array.argsort())
array.sort()
print(array)
print(array.argsort())
print("4 smallest elements are:")
for i in range (0,4):
  print(array[i])
print("5 largest elements are:")
for i in range(-5,0):
  print(array[i])
```

```python

import numpy as np
x=ord('N')+ord('K')
array=np.array([[x,x+50,x+100,x+150,x+200]])
print(x)
print(array)
#using arange
array=np.arange(x,x+250,50).reshape(1,5)
print(array)
tax=((x % 5) + 5) / 100
array2=array+(array*tax)
print("array after adding tax is:")
print(array2)
discount=np.where(array2<x+100,0.05,0.1)
array3=array2-array2*discount
print("array after discount is:")
print(array3)
weekly_sales=np.tile(array3,(3,1))
print("weekly sales is:")
print(weekly_sales)
weekly_sales=weekly_sales+0.2%weekly_sales
print("new weekly sales is:")
print(weekly_sales)
```

```python
import matplotlib.pyplot as plt

x=np.linspace(-10,10,100)
print(x)

```

```python
y=x*x

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("plot of x^2")
plt.grid()
plt.plot(x,y)
plt.show()


```

```python
y=np.sin(x)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("plot of sin(x)")
plt.grid()
plt.plot(x,y)
plt.show()


```

```python
y=np.exp(x)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("plot of exp(X)")
plt.grid()
plt.plot(x,y)
plt.show()
```

```python
y=np.log(abs(x)+1)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("plot of log(abs(x)+1)")
plt.grid()
plt.plot(x,y)
plt.show()
```
